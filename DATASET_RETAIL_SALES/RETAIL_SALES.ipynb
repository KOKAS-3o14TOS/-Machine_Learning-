EVALUAR SUPPLIER Y CAMBIAR LOS 164 ELEMENTOS PARA NO PERDER MUESTRAS, EN CASO DE NO ENCONTRAR CORRELACIÓN ELEMINARLAS YA QUE REPRESENTAN APROXIMADAMENTE UN  0.05466 % DE LAS MUESTRA TOTALES 

```python
# Identificar los valores que no se repiten en la columna 'SUPPLIER'
unique_values = df['SUPPLIER'].value_counts()[df['SUPPLIER'].value_counts() == 1].index.tolist()

# Obtener los índices de las filas con valores faltantes en la columna 'SUPPLIER'
missing_indices = df[df['SUPPLIER'].isna()].index

# Reemplazar los valores faltantes con valores aleatorios de unique_values
for idx in missing_indices:
    df.at[idx, 'SUPPLIER'] = rd.choice(unique_values)

print("DataFrame después de reemplazar los valores faltantes:")
print(df)
```

ELIMINAR EL ÚLTIMO DE LOS VALORES FALTANTE DE ITEM TYPE

```python
d =df[df['ITEM TYPE'].isna()].index
df = df.drop(d)
df.isna().sum()
```

CAMBIAREMOS LOS DATOS POR NUMEROS PARA PODER PROCESAR LA INFORMACIPON. UTILIZAREMOS LA CODIFICACION DE FRECUENCIA 

```python
'''
Codificación de Frecuencia (Frequency Encoding): 
Codifica las categorías según la frecuencia con la que aparecen en el dataset. 
Esto es útil para reducir la dimensionalidad mientras conservas información sobre la distribución.
'''
freq_encoding = df['ITEM DESCRIPTION'].value_counts().to_dict()
df['ITEM DESCRIPTION'] = df['ITEM DESCRIPTION'].map(freq_encoding)

freq_encoding = df['SUPPLIER'].value_counts().to_dict()
df['SUPPLIER'] = df['SUPPLIER'].map(freq_encoding)


q=[]
# Obtiene los valor que no se repitan 
for i in df['ITEM TYPE'].values: 
    if i not in q:
        q.append(i)


item_dict = {item: index + 1 for index, item in enumerate( q)}
df['ITEM TYPE'] = df['ITEM TYPE'].map(item_dict)
df
```

HACEMOS SUBDIVISION DE ELEMENTOS DE TEST, TRAIN Y VALIDACIÓN 

```python
# REVOLVER VALORES O SHUFFLED
# Establecer la semilla para reproducibilidad
np.random.seed(42)

# Mezclar el DataFrame
df= df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
# Calcular el tamaño de los conjuntos
total_size = len(df)
test_size = int(total_size * 0.15)
validation_size = int(total_size * 0.10)
train_size = total_size - test_size - validation_size

# Dividir en entrenamiento, validación y prueba
df_train = df[:train_size]
df_val = df[train_size:train_size + validation_size]
df_test = df[train_size + validation_size:]


```

```python
# ENTRENAMIENTO 75%
# Separar características y variable objetivo
X_train = df_train.drop(columns=['RETAIL SALES'])
y_train = df_train['RETAIL SALES']
```

```python
# VALIDACION 10% 
X_val = df_val.drop(columns=['RETAIL SALES'])
y_val = df_val['RETAIL SALES']
```

```python
# TEST 15%
X_test = df_test.drop(columns=['RETAIL SALES'])
y_test = df_test['RETAIL SALES']
```

```python
print(f"Tamaño del conjunto de entrenamiento: {X_train.shape[0]}")
print(f"Tamaño del conjunto de validación: {X_val.shape[0]}")
print(f"Tamaño del conjunto de prueba: {X_test.shape[0]}")
```
