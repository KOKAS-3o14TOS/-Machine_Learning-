CAMBIAREMOS LOS DATOS POR NUMEROS PARA PODER PROCESAR LA INFORMACION. UTILIZAREMOS LA CODIFICACION DE FRECUENCIA 

```python
'''
Codificación de Frecuencia (Frequency Encoding): 
Codifica las categorías según la frecuencia con la que aparecen en el dataset. 
Esto es útil para reducir la dimensionalidad mientras conservas información sobre la distribución.
'''
freq_encoding = df['ITEM DESCRIPTION'].value_counts().to_dict()
df['ITEM DESCRIPTION'] = df['ITEM DESCRIPTION'].map(freq_encoding)

freq_encoding = df['SUPPLIER'].value_counts().to_dict()
df['SUPPLIER'] = df['SUPPLIER'].map(freq_encoding)


q=[]
# Obtiene los valor que no se repitan 
for i in df['ITEM TYPE'].values: 
    if i not in q:
        q.append(i)


item_dict = {item: index + 1 for index, item in enumerate( q)}
df['ITEM TYPE'] = df['ITEM TYPE'].map(item_dict)
df
```

Multiconealidad

```python

from statsmodels.stats.outliers_influence import variance_inflation_factor

X = sm.add_constant(df_X)
vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]



for num,i in enumerate(vif_data['VIF']):
    if i>10:
        print('Multiconealidad',num,i)
    elif i>5:
        print('Correlación moderada que podría ser preocupante',num,i)
    elif 1<i and i<5:
        print(' Cierta correlación entre la variable independiente,no es la suficiente para ser preocupante',num,i)
    else:
        print('Escenario idela',num,i)
print(vif_data)
```
